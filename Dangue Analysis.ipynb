```python
#imported pandas
import pandas as pd
```

```python
#loaded the csv file into dataframe
a=pd.read_csv("danguedata.csv")
```

```python
#checked first 5 rows of the dataframe
a.head()
```

We can see that all the data are in the same column.So we need to split the column into sperate column

```python
#before splitting , let's rename the column for ease
a.columns=['data']
```

```python
#now let's see either it is  changed or not
a.head()
```

```python
#now let's split the column
a[['Province','Districts','Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sept','Oct','Nov','Grand Total']]=a['data'].str.split(',',expand=True)
```

```python
a.head()
```

Now, we can see that all the comuns are splitted up but now we have to drop the column that we splitted

```python
#dropping the data column
a.drop('data',inplace=True,axis=1)
```

```python
#checking either it is fxed or not
a.head()
```

Now we have another problem , the districs column value have the combination of some code and district name so we will split it

```python
#now spliting the Districts columns into code and District Name
a[['Code','District']]=a['Districts'].str.split(' ',1,expand=True)
```

```python
a.head()
```

Now we can see that the Districts columns is splitted but we still have the column whih we need to drop`

```python
a.drop('Districts',axis=1,inplace=True)
```
